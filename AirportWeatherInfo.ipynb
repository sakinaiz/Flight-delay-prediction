```python
import pandas as pd
import numpy as np

import datetime as dt
```

# load the raw weather data for the selected airports

```python
weather_list = []
df = pd.read_csv('data/climate_data_10airports_2017_mod.csv')
weather_list.append(df)
df = pd.read_csv('data/climate_data_20airports_2017_mod.csv')
weather_list.append(df)

weather_df = pd.concat(weather_list, ignore_index=True)
```

```python
weather_df.columns
```

```python
weather_df.describe()
```

# select required weather information

```python
columns_select = ['CODE','LATITUDE', 'LONGITUDE',
                  'DATE', 'AWND', 'PRCP', 'SNOW']
weather_req_df = weather_df[columns_select]
```

```python
weather_req_df.head(15)
```

```python
weather_req_df['CODE'].nunique()
```

```python
# convert to datetime
date_fmt = '%Y/%m/%d'
weather_req_df.loc[:,'DATE'] = pd.to_datetime(weather_req_df.loc[:,'DATE'], format=date_fmt)
```

```python
weather_req_df.shape
```

```python
weather_req_df['CODE'].value_counts()
```

```python
weather_req_df.isnull().sum()
```

```python
weather_req_df.fillna(-100,inplace=True)
```

# save the weather data

```python
weather_req_df.to_csv('AirportWeatherCleanData2017.csv', index=False)
weather_req_df.to_pickle('AirportWeatherCleanData2017.pkl')
```
